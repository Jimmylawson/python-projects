```python
#Generating a fibonacci sequence
def fibonnacci(n):
    if n <= 1:
        return n
    else:
        return fibonnacci(n - 1) + fibonnacci(n - 2)
```

```python
#valid parentheses
def validParenthenses(s):
    if not s :
        return False
    stack = []
    parenthesis = {
        ")" : "(",
        "]" : "[",
        "}" : "{"
    }

    for char in s:
        if char in parenthesis:
            if stack and stack[-1] == parenthesis[char]:
                stack.pop()
            else:
                return False
        else:
            stack.append(char)

    return True if not stack else False


```

```python
#Min Stack leetcode question
class MinStack:
    def __init__(self):
        self.stack = []
        self.minStack  = []


    def push(self,val):
        self.stack.append(val)
        val = min(val,self.minStack[-1] if self.minStack else val)
        self.minStack.append(val)



    def pop(self):
        self.stack.pop()
        self.minStack.pop()


    def top(self) -> int:
        return self.stack[-1]




    def getMin(self) -> int:

        return self.minStack[-1]


```

```python
def evalRPN(tokens):
        if not tokens:
            return  0

        stack = []

        for c in  tokens:
            if c == '+':
                stack.append(stack.pop() + stack.pop())
            elif c == '-':
                a, b = stack.pop() , stack.pop()
                stack.append(b - a)
            elif c == '*':
                stack.append(stack.pop() * stack.pop())
            elif c == '/':
                a, b = stack.pop() , stack.pop()
                stack.append(int(b / a))
            else:
                stack.append(int(c))

        return stack[0]
```

```python
tokens = ["2","1","+","3","*"]
evalRPN(tokens)
```

```python
#Binary Search
def find_index(nums, target):
    if not nums: return 0

    L, R = 0 , len(nums)

    while L < R:
        mid = (L + R) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] > target:
            R = mid - 1
        else:
            L = mid + 1

    return -1
```

```python
result = find_index([4,8,15,16,27,34,42,46,50],27)
print(result)
```
