```python
#Generating a fibonacci sequence
def fibonnacci(n):
    if n <= 1:
        return n
    else:
        return fibonnacci(n - 1) + fibonnacci(n - 2)
```

```python
#valid parentheses
def validParenthenses(s):
    if not s :
        return False
    stack = []
    parenthesis = {
        ")" : "(",
        "]" : "[",
        "}" : "{"
    }

    for char in s:
        if char in parenthesis:
            if stack and stack[-1] == parenthesis[char]:
                stack.pop()
            else:
                return False
        else:
            stack.append(char)

    return True if not stack else False


```

```python
#Min Stack leetcode question
class MinStack:
    def __init__(self):
        self.stack = []
        self.minStack  = []


    def push(self,val):
        self.stack.append(val)
        val = min(val,self.minStack[-1] if self.minStack else val)
        self.minStack.append(val)



    def pop(self):
        self.stack.pop()
        self.minStack.pop()


    def top(self) -> int:
        return self.stack[-1]




    def getMin(self) -> int:

        return self.minStack[-1]


```

```python
obj = MinStack()
obj.push(-2)
print(obj.stack)
print(obj.minStack)
obj.push(0)
print(obj.stack)
print(obj.minStack)
obj.push(-3)
print(obj.stack)
obj.getMin()
print(obj.minStack)

obj.pop()
print(obj.stack)
print(obj.minStack)
obj.top()
print(obj.stack)
obj.getMin()
print(obj.minStack)
```
