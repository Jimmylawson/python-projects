```python
#Generating a fibonacci sequence
def fibonnacci(n):
    if n <= 1:
        return n
    else:
        return fibonnacci(n - 1) + fibonnacci(n - 2)
```

Input: s = "()[]{}"

Output: true

```python
#valid parentheses
def validParenthenses(s):
    if not s :
        return False
    stack = []
    parenthesis = {
        ")" : "(",
        "]" : "[",
        "}" : "{"
    }

    for char in s:
        if char in parenthesis:
            if stack and stack[-1] == parenthesis[char]:
                stack.pop()
            else:
                return False
        else:
            stack.append(char)

    return True if not stack else False


```
